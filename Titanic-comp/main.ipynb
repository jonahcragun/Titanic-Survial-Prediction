# Titanic Survivor Prediction
model to predict which passengers will survive

```python
# imports
import pandas as pd
```

## Set up training data

```python
# read in training data
t_data = pd.read_csv('train.csv')

# select columns for training
fields = ['Pclass', 'Sex', 'Age', 'Fare']

# create prediction df
X = pd.get_dummies(t_data[fields], columns=['Sex'])

# create predicted value df
y = t_data['Survived']
```

## Split data

```python
#imports 
from sklearn.model_selection import train_test_split
```

```python
# split data to train and test
train_X, val_X, train_y, val_y = train_test_split(X, y)
```

## Train model

```python
# imports
from sklearn.ensemble import RandomForestClassifier
```

```python
# fit model
model = RandomForestClassifier(n_estimator=1000, random_state=42)
model.fit(train_X, train_y)
```

## Make Predictions

```python
predictions = model.predict(val_X)
```

## Evaluate Model

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(predictions, val_y)
print(score)
```

## Tune Model

```python
final_model = RandomForestClassifier(random_state=42)
final_model.fit(X, y)
```

## Use on test data

```python
# read test data
test_data = pd.read_csv('test.csv')

# select test columns
test_X = pd.get_dummies(test_data[fields], columns=['Sex'])

# predict
test_predictions = final_model.predict(test_X)
results = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': test_predictions})
print(results)
```

```python
# write results to csv
results.to_csv('results.csv', index=False)
```
